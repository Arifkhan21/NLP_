```python
import pandas as pd
import numpy as np
import tensorflow as tf
import re
import nltk
```

```python
from gensim.models import word2vec, FastText
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')
data.head(10)
```

```python
stopwords = nltk.corpus.stopwords.words()

def sentence_clean(val):
    regex = re.compile('([^\s\w]|_)+')
    sentence = regex.sub('',val).lower()
    sentence = sentence.split(" ")
    for word in sentence:
        if word in stopwords:
            sentence.remove(word)
    sentence = " ".join(sentence)
    
    return sentence
        
```

```python
data = data.dropna()
for cols in ['question1','question2']:
    data[cols] = data[cols].apply(sentence_clean)
```

```python
data.head(10)
```

```python
corpus = []
for col in ['question1','question2']:
    for sentence in data[col].iteritems():
        sentence = sentence[1].split(" ")
        corpus.append(sentence)
corpus[:2]
```

```python
model = word2vec.Word2Vec(corpus,size=100,workers=4,window=10)
```

```python
model_FT = FastText(corpus,size=100,workers=4,window=10,sg=1,min_count=5)
```
